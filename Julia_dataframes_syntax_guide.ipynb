<b>Calculated Columns

```julia
"""
Assign flag to top rated hospitals
"""
function top_rated_hospitals(rating)
    if typeof(rating) != Float64
        return Missing
    elseif rating >= 4
        return 1
    elseif rating < 4
        return 0
    end
end

# Broadcast top_rated
med_data.top_rated_hospitals = top_rated_hospitals.(med_data.hospital_overall_rating);

# View outputs
first(med_data[:,[:state, :top_rated_hospitals]], 5)
```

<b>Removing methods</b>
* Julia practices multiple dispatch, which I've found to be a little tricky if you're not attentive. It's helpful to know how to remove indivual methods.

```julia
# View methods
println(methods(top_rated_hospitals))

# collect a specific method based on input datatype
m = @which top_rated_hospitals(Any)

# View selected method
println(m)

# Remove method
Base.delete_method(m)

# View methods
println(methods(top_rated_hospitals))
```
